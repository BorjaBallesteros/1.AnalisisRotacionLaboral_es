#### · Mapa de calor para identificar las correlacciones

```python
correlaciones = df.corr()
sns.heatmap(correlaciones, annot=True, cmap="coolwarm")
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

Ajustamos el tamaño para ver mejor los resultados y creamos una máscara para ver solo los valores por encima de 0.5

```python
threshold = 0.5
mascara = abs(correlaciones) < threshold

plt.figure(figsize=(20, 8))
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f", mask=mascara)
plt.title("Mapa de Correlaciones (|Correlación| > 0.5)")
plt.show()
```

# Conclusiones del analisis realizado
## 1. Los factores asociados con rotación de los empleados son los siguientes:
- **Los salarios bajos contribuyen a la rotación:** ya que presentan una mayor tasa de rotación en comparación con salarios más altos.
- **La antigüedad en la empresa:** la rotación del personal es más alta en los empleados que tienen menos años en la empresa.
- **Horas de trabajo:** los empleados que reportan no hacer extras tienen una mayor probabilidad de abandonar la empresa.
- **Satisfacción laboral:** la satisfacción laboral es más baja en los empleados que renuncian. Esto indica que la forma en la que se percibe el ambiente laboral influye en la decisión de dejar o no la empresa.
- **Conciliación entre la vida laboral y la personal:** las mayores tasas de rotación se encuentran entre los perfiles con una menor puntuación en esta variable, lo que evidencia la importancia de un buen equilibrio entre el trabajo y la vida personal.
## 2. Patrones identificados en la rotación de personal:
- En el mapa de calor se puede identificar que **el ingreso mensual, los años en la empresa y la satisfacción laboral** están relaccionados con la retención de los empleados.
- Los empleados que no realizan horas extras tienen una menor tasa de rotación, por lo que las **cargas de trabajo y la flexibilidad** podrían ser factores determinantes para evitar las fugas de talento.
- El porcentaje de rotación en la empresa es del **16.12%**, lo que indica que aunque un gran porcentaje de los empleados permanecen en la empresa, existe un gran número de personas que deciden renunciar.
## 3. Recomendaciones para mejorar la retención de talento:
- **Revisión de políticas salariales:** implementar ciertas estrategias de ajuste salarial dirigidas a los empleados con menores ingresos.
- **Flexibilidad laboral:** identificar y reducir las cargas de trabajo excesivas y ofrecer flexibilidad en los horarios o teletrabajo.
- **Incentivos y beneficios adicionales:** estudiar la posibilidad de ofrecer bonos por desempeño, días libres adicionales o programas que puedan mejorar la conciliación entre la vida personal y la laboral.
- **Mejora del ambiente laboral:** implementar encuestas de satisfacción laboral para identificar y atajar problemas que puedan afectar a la motivación de la plantilla.

Todas estas conclusiones pueden servir como punto de partida para desarrollar estrategias que reduzcan la rotación y el coste que ello ocasiona y mejorar la retención de talento en la empresa.
